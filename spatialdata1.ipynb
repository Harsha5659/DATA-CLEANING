```python
import folium

# Define the initial map coordinates
map_center = [16.1994785, 77.319139]  # Replace with your desired location

# Create a new map object
tourist_map = folium.Map(location=map_center, zoom_start=12)

# Define the tourist locations with their respective details
tourist_locations = [
    {
        'name': 'Raichur Fort',
        'location': [16.1994785, 77.319139],
        'image': 'https://traveltalesofafernweh.files.wordpress.com/2017/02/img_9479sd.jpg?w=1080',
        'description': 'A historic fort in Raichur, Karnataka, India.'
    },
    {
         'name': 'Maliabad Fort',
        'location': [16.1739041,77.2115246],
        'image': 'https://pbs.twimg.com/media/D3ec6y2UIAAmUGt.jpg',
        'description': 'historical monument.'
    },
    {'name': 'Mantralayam',
        'location': [15.9415801,77.4293952],
        'image': 'https://cdn.triphippies.com/wp-content/uploads/2023/09/Mantralayam-Temple-1.webp',
        'description': 'Mantralayam for that healing touch.'
        }
    # Add more tourist locations here
]

# Add markers and popups to the map
for location in tourist_locations:
    folium.Marker(
        location=location['location'],
        tooltip=location['name'],
        icon=folium.Icon(icon='info-sign')
    ).add_to(tourist_map)

    folium.Marker(
        location=location['location'],
        popup=folium.Popup(
            f"<h3>{location['name']}</h3><img src='{location['image']}' width='200'><p>{location['description']}</p>",
            max_width=300
        ),
        icon=folium.Icon(color='green')
    ).add_to(tourist_map)

# Display the map
tourist_map
```
