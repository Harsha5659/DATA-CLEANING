```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
mycolors=["red","yellow","pink","green"]
mylabels=["cherries","banana","strawberry","grapes"]
d=[0.3,0.4,0.5,0.2]
plt.pie(x,labels=mylabels,colors=mycolors,explode=d,shadow=True)
plt.show()
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
plt.barh(x,y,color=mycolors)
mycolors=["red","yellow","pink","green"]
plt.title("Marks",fontsize=30)
plt.xlabel("x axis",fontsize=15)
plt.ylabel("y axis",fontsize=15)
plt.show()
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
mycolors=["r","y","b","g"]
mysizes=[100,200,300,400]
plt.scatter(x,y,c=mycolors,s=mysizes,alpha=0.9,marker="*",edgecolor="black")
plt.title("scatter",fontsize=30)
plt.xlabel("x axis",fontsize=15)
plt.ylabel("y axis",fontsize=15)
plt.show()
```
